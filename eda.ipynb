```python
columns = {

    'timestamp': 'Time',
    'cnt': 'Count',
    't1': 'Temperature',
    't2': 'Feels like Temperature',
    'hum' : 'Humidity',


}

df.rename(columns=columns, inplace=True)
```

```python
df.head()
```

```python
df.Humidity = df.Humidity / 100
df
```

```python
season = {
    '0.0' : 'spring',
    '1.0' : 'Summer',
    '2.0' : 'Fall',
    '3.0' : 'Winter'
}
df.season = df.season.astype('str')
df.season = df.season.map(season)
df.head()
```

```python
df.weather_code.unique()
```

```python
weather = {
    '1.0': 'Clear',
    '2.0': 'Scattered Clouds',
    '3.0': 'Broken Clouds',
    '4.0': 'Cloudy',
    '7.0': 'Rain',
    '10.0': 'Thundersto',
    '26.0': 'Snow'
}
df.weather_code = df.weather_code.astype('str')
df.weather_code = df.weather_code.map(weather)
df.head()
```

```python
df.rename(columns={'weather_code': 'weather'})
```

```python
df.to_excel('bikes_final.xlsx', sheet_name='Data')
```
